```python
import pandas as pd

#create dataset
df = pd.DataFrame({'age': [1, 2, 4, 5, 5, 6, 6, 7, 8, 10, 11, 11, 12, 12, 14],
                   'review': [64, 66, 76, 73, 74, 81, 83, 82, 80, 88, 84, 82, 91, 93, 89]})
```

```python
import matplotlib.pyplot as plt

plt.scatter(df.hours, df.score)
plt.title('Age vs. Review')
plt.xlabel('Hours')
plt.ylabel('Score')
plt.show()
```

```python
df.boxplot(column=['age'])
```

```python
import statsmodels.api as sm

#define response variable
y = df['age']

#define explanatory variable
x = df[['review']]

#add constant to predictor variables
x = sm.add_constant(x)

#fit linear regression model
model = sm.OLS(y, x).fit()

#view model summary
print(model.summary())
```
